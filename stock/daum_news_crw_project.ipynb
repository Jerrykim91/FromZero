```python
def PageUrl( txt,page ):
    """
    # Page assignment
    from urllib.parse import quote
    """
    keyword    = quote(txt) # 쿼리 
    fstPage = f'https://search.daum.net/search?nil_suggest=btn&w=news&DA=SBC&cluster=y&q={keyword}'
    
    if page == 1 :
        #  page data
        return fstPage

    else:
        # Non-page data
        otherPage = fstPage + f'&p={page}'
        return otherPage
```

```python
import re, string
import requests 
import lxml.html
import pandas as pd 
import sqlite3
from pandas.io import sql
import os, time


page = 1
max_page = 0
tit = []
txt = []
url = [] 


# 함수 
def getDetail(url):
    """
    """
    body = list()
    punc = '[!"#$%&\'()*+,-./:;<=>?[\]^_`{|}~“”·]'
    response = requests.get(url)
    root = lxml.html.fromstring(response.content)
    
    try: #페이지에서 p태그 다 못가지고오는 경우가 있어서 추가작성 
        for i in range(100): # 임의로 숫자 지정 
            for p in root.xpath(f'//*[@id="harmonyContainer"]/section/p[{i}]'):
                if p.text:
                    body.append(re.sub(punc,'',p.text))
        fullBody = ' '.join(body)
        return fullBody
    except:
        try:
            for p in root.xpath(f'//*[@id="harmonyContainer"]/section/p'):
                if p.text:
                    body.append(re.sub(punc,'',p.text))
            fullBody = ' '.join(body)
            return fullBody
        except:
            print(url)
            


running = True 
while running: 

    url = f'https://news.daum.net/breakingnews/digital?page={page}&regDate={REG_DATE}'
    response = requests.get(url)
    root = lxml.html.fromstring(response.content)
    cnt = 1
    for li in root.xpath('//*[@id="mArticle"]/div[3]/ul/li'):
        a = li.xpath('div/strong/a')[0]
        title   = a.text
        url     = a.get('href')
        content = getDetail(url)
        print(page, title) # 페이지와 제목 표출 -> 1페이지 10개 ,2페이지 10개 
    #     print(title, url, content)

        tit.append(title)
        txt.append(content)
    
    for a in root.xpath('//*[@id="mArticle"]/div[3]/div/span/a'):
        try:
            num = int(a.text)
            if max_page < num :
                max_page = num
        except:
            pass
        
    
    span = root.xpath('//*[@id="mArticle"]/div[3]/div/span/a[@class="btn_page btn_next"]')
    if (len(span) <= 0 ) & ( page > max_page ):
        break
    else:
        page = page + 1
    print(num) # 마지막 페이지
    time.sleep(1)      
```
