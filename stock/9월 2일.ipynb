```python
def scrape_list_page(response):
    root = lxml.html.fromstring(response.content)
    root.make_links_absolute(response.url)
    for a in root.cssselect('.view_box .book_tit a'):
        url = a.get('href')
        # yield 구문으로 제너레이터의 요소 반환
        yield url
```

```python
main2()
```

```python
# 상세 페이지 스크래핑
def scrape_detail_page(response):
    
    root = lxml.html.fromstring(response.content)
    
    ebook = {                        # 딕셔너리 타입으로 저장
        'url' : response.url,
        'title' : root.cssselect('.store_product_info_box h3')[0].text_content(),
        'price' : root.cssselect('.pbr strong')[0].text_content(),
        'content':[normalize_spaces(p.text_content()) # 공백 줄이기
                  for p in root.cssselect('#tabs_3 .hanbit_edit_view p')
                   if normalize_spaces(p.text_content()) != '']  # 길이가 0인 문자열이 아닌애들만 필터링              
    }
    return ebook


def normalize_spaces(s):
    return re.sub(r'\s+','',s).strip()
```

```python
import time
import re
def main():
    #여러 페이지에서 크롤링을 위해 Session 사용
    session = requests.Session()
    #scrape_list_page() 함수를 호출해서 제너레이터를 추출
    response = session.get('https://www.hanbit.co.kr/store/books/new_book_list.html')
    urls= scrape_list_page(response)
    print(urls)
    
    for url in urls:
        time.sleep(1) # 1초간 대기 , 안쓰면 사이트에서 차단할수도 .. 
        response = session.get(url) # Session을 사용해 상세 페이지를 추출
        ebook = scrape_detail_page(response)
        print(ebook) # 상세 정보 출력
        #break # 주석을 활용해 체크 후 출력
main()
```

```python
page=1
max_page=0
tit =[]
txt=[]
url =[]

def get_detail(url):
    body = []
    punc = '[!"#$%&\'()*+,-./:;<=>?[\]^_{|}~"",]'
    response = requests.get(url)
    root = lxml.html.fromstring(response.content)


# 페이지 번호 중에서 max 페이지 가져오기
    
    
    
while(True):       
    
    url = "https://news.daum.net/breakingnews/digital?page={}&regDate={}".format(page,REG_DATE)
    response = requests.get(url)
    root =lxml.html.fromstring(response.content)
    for li in root.xpath('//*[@id="mArticle"]/div[3]/ul/li'):
        a=li.xpath('div/strong/a')[0]
        title=a.text
        url=a.get('href')
        content=get_detail(url)
        print(page,title)
        
        tit.append(title)
        txt.append(content)
        
    
    for a in root.xpath('//*[@id="mArticle"]/div[3]/div/span/a'):
        try:
            num = int(a.text)
            if max_page<num:
                max_page = num
               
        except:
            pass
    
# 마지막 페이지 여부 확인
    span = root.xpath('//*[@id="mArticle"]/div[3]/div/span/a[@class="btn_page btn_next"]')

    if (len(span) <=0) & (page > max_page):
        break
    else:
        page =page+1
        
        
    time.sleep(1)    
```
